```python
## 실시간 검증 ##
실시간 테스트(사용자 입력)가 되어서 실시간 시스템 결과물(LLM + RAG)이 나올 수 있는지를 검증하기 위한 내용입니다. 실시간으로 pubmed 검색을 지원하는 만큼, 다소 시간은 소요되나 여러 단어를 돌려보았을 때 꽤 좋은 성능을 보이는 것을 확인했습니다. 특히 disease->gene 검색보다 gene->disease 검색에서 좋은 성능을 보입니다.  
```

```python
import sys
from pathlib import Path

ROOT = Path("..").resolve()
sys.path.append(str(ROOT))

from src.llm import answer_with_rag_llm
```
